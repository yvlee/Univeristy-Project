```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
cityPairs = pd.read_csv('CityPairs.csv', sep = ',')
cityPairs
```

```python
cityPairs = cityPairs[(cityPairs['AustralianPort'] == 'Sydney') | (cityPairs['AustralianPort'] == 'Melbourne') | (cityPairs['AustralianPort'] == 'Brisbane') | (cityPairs['AustralianPort'] == 'Perth')] 
cityPairs
```

```python
total = cityPairs.groupby(['Year','AustralianPort'])['Passengers_Total','Freight_Total_(tonnes)','Mail_Total_(tonnes)']
cityPairs2 = total.sum()
cityPairs2
```

```python
cityPairs2.to_csv('output.csv')
```
