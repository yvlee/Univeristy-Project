```python
from matplotlib import pyplot as plt
import pandas as pd
df = pd.read_csv('CityPairs.csv', sep = ',')
df
```

```python
df1 = df[(df['Country'] == 'USA') | (df['Country'] == 'China')]
df1
```

```python
df1 = df1.groupby(['Year','Country'], as_index=False)['Passengers_Total'].sum()
df1
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
df1.boxplot(column = 'Passengers_Total', by = 'Country')
plt.show()
```
