```python
from matplotlib import pyplot as plt
import pandas as pd
df = pd.read_csv('CityPairs.csv', sep = ',')
df
```

```python
df1 = df[df['AustralianPort'] == 'Sydney']
df1
```

```python
df1 = df.groupby(['Year'], as_index=False)['Passengers_Total'].sum()
df1
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(df1['Year'], df1['Passengers_Total'])
plt.xlabel('Year')
plt.ylabel('Passengers_Total')
plt.show()
```

```python
outlierRemove = df1[df1['Year'] < 2016]
outlierRemove 
```

```python
from scipy.stats import linregress

slope, intercept, r_value, p_value, std_err = linregress(outlierRemove['Year'],outlierRemove['Passengers_Total'])
line = [slope*xi + intercept for xi in outlierRemove['Year']]
plt.plot(outlierRemove['Year'],line,'r-', linewidth=3)
plt.scatter(outlierRemove['Year'], outlierRemove['Passengers_Total'])
plt.show()
```

```python
newCond = df1[(df1['Year'] < 2016) & (df1['Year'] > 2003)]
newCond
```

```python
from scipy.stats import linregress

slope, intercept, r_value, p_value, std_err = linregress(newCond['Year'],newCond['Passengers_Total'])
line = [slope*xi + intercept for xi in newCond['Year']]
plt.plot(newCond['Year'],line,'r-', linewidth=3)
plt.scatter(newCond['Year'], newCond['Passengers_Total'])
plt.show()
```
