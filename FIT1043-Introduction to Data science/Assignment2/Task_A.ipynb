```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('CityPairs.csv', sep = ',')
df = df[df['Country'] == 'USA']
df
```

```python
df1 = df.groupby(['Month_num'], as_index=False)['Mail_In_(tonnes)','Mail_Out_(tonnes)'].mean()
df1
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(df1['Month_num'], df1['Mail_In_(tonnes)'])
plt.xlabel('Month')
plt.ylabel('Mail_In_(tonnes)')
plt.show()
```

```python
plt.scatter(df1['Month_num'], df1['Mail_Out_(tonnes)'])
plt.xlabel('Month')
plt.ylabel('Mail_Out_(tonnes)')
plt.show()
```

```python
df1.boxplot(column = ['Mail_In_(tonnes)','Mail_Out_(tonnes)'])
plt.show()

#df1.boxplot(column = 'Month_num', by = 'Mail_Out_(tonnes)')
```

```python
filt = df1[((df1['Mail_In_(tonnes)'] < 13.8) & (df1['Mail_Out_(tonnes)'] < 7))]
filt.boxplot(column = ['Mail_In_(tonnes)','Mail_Out_(tonnes)'])
plt.show()
```
