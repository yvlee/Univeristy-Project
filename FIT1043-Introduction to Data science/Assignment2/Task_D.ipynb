```python
from matplotlib import pyplot as plt
import pandas as pd
tsunami_data = pd.read_csv('sources.csv', sep = ',')
tsunami_data.head()
```

```python
tsunami_data.dtypes
tsunami_data
```

```python
tsunami_data['COUNTRY_FREQUENCY'] = tsunami_data.groupby(tsunami_data.COUNTRY)['YEAR'].transform('count')
tsunami_data.head()
```

```python
tsunami_data.COUNTRY[tsunami_data.COUNTRY_FREQUENCY > 50].value_counts().plot(kind='bar',
                                             legend=False,
                                             figsize=(12,5),
                                             title="Tsunamis frequency by countries > 50",
                                             fontsize=12,
                                             alpha=0.5
                                            );
```

```python
tsunami_data1 = tsunami_data[((tsunami_data['COUNTRY'] == 'JAPAN') |(tsunami_data['COUNTRY'] == 'MEXICO')) & (tsunami_data['YEAR'] >1999)]
tsunami_data1
```

```python
freq = tsunami_data1.COUNTRY.value_counts()
freq
```

```python
tsunami_data1 = tsunami_data1.groupby(['COUNTRY','YEAR','COUNTRY_FREQUENCY'], as_index=False)['PRIMARY_MAGNITUDE'].mean()
tsunami_data1
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
tsunami_data1.boxplot(column = 'PRIMARY_MAGNITUDE', by = 'COUNTRY')
plt.show()
```
